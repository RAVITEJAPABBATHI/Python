4.Write a Python program to find the volume of a sphere with diameter 12 cm.
Formula: V=4/3 *  * r
3

```python
import math
```

```python
def getSphereVolume(d):
    r=d/2
    return (4/3)*math.pi*(r**3)
print(" Volume of the Sphere with diameter 12 is {}".format(getSphereVolume(12)))
```
